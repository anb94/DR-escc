```python
import pandas as pd
import os
import numpy as np
```

```python
egfr_c5mf = "/Users/anb/Desktop/egfr_c5mf.tsv"
egfr_c5bp = "/Users/anb/Desktop/egfr_c5bp.tsv"
egfr_c5cc = "/Users/anb/Desktop/egfr_c5cc.tsv"

egfr_c5mf_df = pd.read_csv(
    egfr_c5mf, comment="#", header=0, sep='\t', low_memory=False
)


egfr_c5bp_df = pd.read_csv(
    egfr_c5bp, comment="#", header=0, sep='\t', low_memory=False
)

egfr_c5cc_df = pd.read_csv(
    egfr_c5cc, comment="#", header=0, sep='\t', low_memory=False
)
```

```python
egfr_c5mf_df_pval05_ESneg = egfr_c5mf_df_pval05[egfr_c5mf_df_pval05['ES'] <= 0]
egfr_c5bp_df_pval05_ESneg = egfr_c5bp_df_pval05[egfr_c5bp_df_pval05['ES'] <= 0]
egfr_c5cc_df_pval05_ESneg = egfr_c5cc_df_pval05[egfr_c5cc_df_pval05['ES'] <= 0]
```
