```python
import os
import pandas as pd
import numpy as np
```

```python
# DEG file

DEG_file = "/Users/anb/Documents/CMEB-Lab/ESCC-Drug-Repositioning/DR-escc-data/output/GSE23400//filtered_de_results_1fc_05pval.tsv"

DEGs = pd.read_csv(
    DEG_file, comment="#", header=0, sep='\t', low_memory=False
)
```

```python
DEGS_mini = DEGs[['Gene.Symbol', 'logFC', 'P.Value']]
```

```python
DEGS_mini
```

```python
DEGS_mini.sort_values(['Gene.Symbol', 'abslogfc', 'P.Value'], ascending=False, inplace=True)
DEGS_mini.drop_duplicates(subset = ["Gene.Symbol"], inplace=True)
```
