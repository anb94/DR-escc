```python
import os
import pandas as pd
import numpy as np
```

```python
# DEG file

DEG_file = "/Users/anb/Documents/CMEB-Lab/ESCC-Drug-Repositioning/DR-escc-data/output/GSE23400/filtered_de_results_1fc_05pval.tsv"

DEGs = pd.read_csv(
    DEG_file, comment="#", header=0, sep='\t', low_memory=False
)
```

```python
DEGs
```

```python
DEGs['Gene.Symbol-drop-slash'] = DEGs['Gene.Symbol'].str.split(' ').str[0]
```

```python
DEGs
```

```python
DEGs['Gene.Symbol-drop-slashdash'] = DEGs['Gene.Symbol-drop-slash'].str.split('-').str[0]
```

```python
DEGs
```

```python
DEGs['Gene.Symbol'] = DEGs['Gene.Symbol-drop-slashdash'].copy()
```

```python
DEGs
```

```python
DEGS_mini = DEGs[['Gene.Symbol', 'logFC', 'P.Value']]
```

```python
DEGS_mini.sort_values(['Gene.Symbol', 'P.Value'], ascending=True, inplace=True)
DEGS_mini.drop_duplicates(subset = ["Gene.Symbol"], inplace=True)



DEGs.sort_values(['Gene.Symbol', 'P.Value'], ascending=True, inplace=True)
DEGs.drop_duplicates(subset = ["Gene.Symbol"], inplace=True)
```

```python
DEGS_mini
```

```python
DEGS_mini.sort_values(['P.Value'], ascending=True, inplace=True)

DEGs.sort_values(['P.Value'], ascending=True, inplace=True)
```

```python
DEGS_mini
```

```python
DEGs
```

```python
DEGS_mini.reset_index(inplace=True,drop=True)
```

```python
out_dir = "/Users/anb/Documents/CMEB-Lab/ESCC-Drug-Repositioning/DR-escc-data/output/GSE23400/DEG-analysis"

DEGS_mini.to_csv(os.path.join(out_dir, "DEGs_mini_sorted_pval.tsv"), sep='\t', index=False)
DEGS_mini[['Gene.Symbol']].to_csv(os.path.join(out_dir, "DEGs_mini_sorted_pval_genes.tsv"), sep='\t', index=False)

DEGs.to_csv(os.path.join(out_dir, "filtered_de_results_1fc_05pval_dupsdropped.tsv"), sep='\t', index=False)
DEGs[['Gene.Symbol']].to_csv(os.path.join(out_dir, "filtered_de_results_1fc_05pval_dupsdropped_genes.tsv"), sep='\t', index=False)
```
