```python
filepath = "/Users/anb/Desktop/GSE44021_RAW"
ID_list = os.listdir(filepath)
```

```python
for i in range(len(ID_list)):
    ID_list[i] = ID_list[i].replace('.CEL.gz','')
    p = ID_list[i].split('_')
    ID_list[i] = p[1]
```

```python
group = pd.DataFrame()
```
