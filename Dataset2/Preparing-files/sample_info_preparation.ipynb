```python
import pandas as pd
import os
```

```python
sample_info = "/Users/anb/Desktop/sampleinfo.tsv"

sample_info_df = pd.read_csv(
    sample_info, comment="#", header=0, sep='\t', low_memory=False
)
```

```python
sample_info_df
```

```python
group = pd.DataFrame()
group['sample'] = sample_info_df['title']
group['case'] = 0
group['ctrl'] = 0
group['group'] = 0
```

```python
group.loc[group['sample'].str.contains("N")==True,'ctrl'] = 1
group.loc[group['sample'].str.contains("T")==True,'case'] = 1

group.loc[group['sample'].str.contains("N")==True,'group'] = "normal"
group.loc[group['sample'].str.contains("T")==True,'group'] = "tumour"
```

```python
group
```

```python
group.loc[group['sample'].str.contains("AE01")==True]
```

```python
group.sort_values('sample', axis=0, ascending=True, inplace=True)
```

```python
group
```

```python
group.reset_index(inplace=True, drop=True)
```

```python
group
```

```python
a=[(x+1) for x in range(113) for _ in range(2)]
b=pd.DataFrame(columns=['number'])
b['number'] = a
b.reset_index
b
```

```python
group['patient'] = b['number'].copy()
```

```python
group
```

```python
sampleInfoCorrected = pd.merge(left=sample_info_df, right=group, how='left', left_on='title', right_on='sample')
```

```python
out_dir = "/Users/anb/Desktop/"

sampleInfoCorrected.to_csv(os.path.join(out_dir, "GSE23400_sampleinfoCorrected.tsv"), sep='\t', index=False, header=True)
```
