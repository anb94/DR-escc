```python
import pandas as pd
import os
```

```python
egfr_protiens_file = "/Users/anb/Desktop/egfr_proteins_copy.tsv"

egfr_protiens = pd.read_csv(
    egfr_protiens_file, comment="#", header=0, sep='\t', low_memory=False
)
```

```python
egfr_protiens[egfr_protiens['proteins'] == 'CRK']
```

```python
DEG_file = "/Users/anb/Desktop/unfiltered_de_results.csv"

DEG_df = pd.read_csv(
    DEG_file, comment="#", header=0, sep=',', low_memory=False
)
```

```python
DEG_df
```

```python
tempdf =  pd.merge(left=egfr_protiens, right=DEG_df, how='inner', left_on='proteins', right_on='Gene.Symbol')
```

```python
tempdf
```

```python
out_dir = "/Users/anb/Desktop/"

tempdf.to_csv(os.path.join(out_dir, "egfr-proteins-DEG-unfiltered.tsv"), sep='\t', index=False, header=True)
```

```python
upreg = tempdf[tempdf['logFC'] >= 0]
downreg = tempdf[tempdf['logFC'] <= 0]
```

```python
out_dir = "/Users/anb/Desktop/"

upreg['Gene.Symbol'].to_csv(os.path.join(out_dir, "upreg-egfr-proteins-DEG-unfiltered.tsv"), sep='\t', index=False, header=True)
downreg['Gene.Symbol'].to_csv(os.path.join(out_dir, "downreg-egfr-proteins-unfiltered.tsv"), sep='\t', index=False, header=True)
```
